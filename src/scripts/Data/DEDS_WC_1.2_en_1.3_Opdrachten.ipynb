## Selecties op meerdere tabellen zonder functies

Geef een overzicht met daarin het eerste adres en de stad van vestigingen van klanten uit ‘Netherlands’ (2 kolommen, 20 rijen) 

```python
sql_query = "SELECT * FROM country"
sql_query1 = "SELECT * FROM retailer_site"
table_sales = pd.read_sql_query(sql_query1, sales_conn)
country = pd.read_sql_query(sql_query, sales_conn)

merge = pd.merge(table_sales, country, left_on='COUNTRY_CODE', right_on='COUNTRY_CODE', how='inner')
nederlands = merge['COUNTRY'] == 'Netherlands'
merge.loc[nederlands, ['ADDRESS1', 'CITY']]
```

Geef een overzicht met daarin de productnamen die tot het producttype ‘Eyewear’ behoren. (1 kolom, 5 rijen) 

```python
sql_query = "SELECT * FROM product"
sql_query1 = "SELECT * FROM product_type"
product = pd.read_sql_query(sql_query, sales_conn)
product_type = pd.read_sql_query(sql_query1, sales_conn)

merge = pd.merge(product, product_type, left_on='PRODUCT_TYPE_CODE', right_on='PRODUCT_TYPE_CODE', how='inner')
eyewear = merge['PRODUCT_TYPE_EN'] == 'Eyewear'
merge.loc[eyewear, ['PRODUCT_NAME']]
```

Geef een overzicht met daarin alle unieke eerste adressen van klantvestigingen en de voornaam en achternaam van de verkopers die ‘Branch Manager’ zijn en aan deze vestigingen hebben verkocht (3 kolommen, 1 rij) 

```python
sql_query = "SELECT * FROM retailer_site"
sql_query1 = "SELECT * FROM order_header"
sql_query2 = "SELECT * FROM sales_staff"

retailer_site = pd.read_sql_query(sql_query, sales_conn)
order_header = pd.read_sql_query(sql_query1, sales_conn)
sales_staff = pd.read_sql_query(sql_query2, sales_conn)


merge = pd.merge(retailer_site, order_header, left_on='RETAILER_SITE_CODE', right_on='RETAILER_SITE_CODE', how='inner')
merge = pd.merge(merge, sales_staff, left_on='SALES_STAFF_CODE', right_on='SALES_STAFF_CODE', how='inner')
conditions = merge['POSITION_EN'] == 'Branch Manager'
merge.loc[conditions, ['ADDRESS1', 'FIRST_NAME', 'LAST_NAME']].drop_duplicates()
```

Geef een overzicht met daarin van de verkopers hun functie en indien zij iets hebben verkocht de datum waarop de verkoop heeft plaatsgevonden. Laat alleen de verschillende namen van de posities zien van de verkopers die het woord ‘Manager’ in hun positienaam hebben staan. (2 kolommen, 7 rijen) 

```python
merge = pd.merge(sales_staff, order_header, left_on='SALES_STAFF_CODE', right_on='SALES_STAFF_CODE', how='outer')
condition = merge['POSITION_EN'].str.contains('Manager')
merge.loc[condition, ['POSITION_EN', 'ORDER_DATE']].drop_duplicates()
```

Geef een overzicht met daarin de verschillende namen van producten en bijbehorende namen van producttypen van de producten waarvoor ooit meer dan 750 stuks tegelijk verkocht zijn. (2 kolommen, 9 rijen) 

```python
sql_query3 = "SELECT * FROM order_details" 
order_details = pd.read_sql_query(sql_query3, sales_conn)

merge = pd.merge(order_details, product, on='PRODUCT_NUMBER', how='inner')
merge = pd.merge(merge, product_type, on='PRODUCT_TYPE_CODE', how='inner')
condition = merge['QUANTITY'] > 750
merge.loc[condition, ['PRODUCT_NAME', 'PRODUCT_TYPE_EN']].drop_duplicates()
```

Geef een overzicht met daarin de productnamen waarvan ooit meer dan 40% korting is verleend. De formule voor korting is: (unit_price - unit_sale_price) / unit_price (1 kolom, 8 rijen) 

```python
product_order = pd.merge(product, order_details, on='PRODUCT_NUMBER', how='inner')
condition = ((product_order['UNIT_PRICE'] - product_order['UNIT_SALE_PRICE']) / product_order['UNIT_PRICE']) > 0.4
product_order.loc[condition, ['PRODUCT_NAME']].drop_duplicates()
```

Geef een overzicht met daarin de retourreden van producten waarvan ooit meer dan 90% van de aangeschafte hoeveelheid is teruggebracht (return_quantity/quantity). (1 kolom, 3 rijen) 

```python
sql_query4 = "SELECT * FROM returned_item" 
returned_item = pd.read_sql_query(sql_query4, sales_conn)
sql_query5 = "SELECT * FROM return_reason" 
return_reason = pd.read_sql_query(sql_query5, sales_conn)

merge = pd.merge(returned_item, order_details, on='ORDER_DETAIL_CODE', how='inner')
merge = pd.merge(merge, return_reason, on='RETURN_REASON_CODE', how='inner')
condition = (merge['RETURN_QUANTITY'] / merge['QUANTITY']) > 0.9
merge.loc[condition, ['RETURN_DESCRIPTION_EN']].drop_duplicates()
```

Geef met behulp van bovenstaande hulpcode een overzicht met daarin op basis van het aantal jaar dat iemand in dienst is of een medewerker ‘kort in dienst’ (minder dan 25 jaar in dienst) of een ‘lang in dienst’ (groter gelijk dan 12 jaar in dienst) is. Geef daarbij per medewerker in een aparte kolom zowel ‘kort in dienst’ als ‘lang in dienst’ aan. Gebruik (wederom) een for-loop.<br>
(2 kolommen, 102 rijen) 

```python
from datetime import datetime
for index, row in sales_staff.iterrows():
    date_hired = sales_staff.at[index, 'DATE_HIRED']
    date_format = '%Y-%m-%d'
    date_obj = datetime.strptime(date_hired, date_format)

    verschil = year_now - date_obj.year

    if verschil < 12:
        sales_staff.at[index, 'Diensttijd'] = 'Kort in dienst'
    else:
        sales_staff.at[index, 'Diensttijd'] = 'Lang in dienst'
sales_staff.loc[:,['LAST_NAME', 'Diensttijd']]
```
